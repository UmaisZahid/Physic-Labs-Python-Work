### Operational Amplifier Data

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(13.5,9)) 
Freq10, Gain10, Gain_Uncertainty10 = np.loadtxt("Op Amp Data - 10 Gain.csv", unpack=True, delimiter=",")
Freq100, Gain100, Gain_Uncertainty100 = np.loadtxt("Op Amp Data - 100 Gain.csv",unpack=True, delimiter=",")
Freq1000, Gain1000, Gain_Uncertainty1000 = np.loadtxt("Op Amp Data - 1000 Gain.csv", unpack=True, delimiter=",")

plt.errorbar(Freq10, Gain10,yerr=Gain_Uncertainty10,fmt='go', label="$\mathit{Gain\  of \ 10 \ (R1=100000) \ (R2=10000) }$", linewidth=1.25)
plt.errorbar(Freq100, Gain100,yerr=Gain_Uncertainty100,fmt='bo', label="$\mathit{Gain \  of \ 100 \ (R1=1000000) \ (R2=10000) }$", linewidth=1.25)
plt.errorbar(Freq1000, Gain1000,yerr=Gain_Uncertainty1000,fmt='ro', label="$\mathit{Gain \  of \ 1000 \ (R1=1000000) \ (R2=1000) }$", linewidth=1.25)

plt.legend(loc="best")

plt.yscale('log')
plt.xscale('log')

plt.xlim([-1000,10000000])
plt.ylim([-1000,1500])


plt.title('$\mathrm{Experimental \ Gain \ Against \ Frequency \ of \ Input \ Signal}$',loc="center", fontsize=20)   #Add centred title as Roman font in Math Text 
plt.ylabel('$\mathit{Log(Frequency) \ (Hz)}$',fontsize=15)   #Add labels as Math Text using $ 
plt.xlabel('$\mathit{Log(Gain)}$', fontsize=15)

plt.savefig("Experimental Gain Against Frequency of Input Signal", dpi=300)
```
