## Fields and waves plot

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

r = np.linspace(0,2, 1000)
plt.figure(figsize=(13.5,9)) 
plt.grid(color='r', linestyle='-', linewidth=0.2)
naturalFrequency = 2


def plotFrequencyFunction(x):
    magnificationRatio = 1/np.sqrt(np.square((1-np.square(r)))+np.square(2*r*x))
    maxMagIndex = np.argmax(magnificationRatio)
    resonanceFrequencyRatio = r[maxMagIndex]
    qualityFactor = resonanceFrequencyRatio/(2*x)
    print(qualityFactor)
    plt.plot(r, magnificationRatio, label="$\mathit{Damping \ ratio \ }$"+str(x), linewidth=1)
    
    
for x in {0.1,0.15,0.3,0.2,0.25,0.5,0.6,0.9,1,10}:
    plotFrequencyFunction(x)
    plt.legend(loc="best")

plt.title('$\mathrm{Frequency \ Ratio \ Against \ Magnification \ Factor}$',loc="center", fontsize=30)   
plt.xlabel('$\mathit{Frequency \ Ratio \ (Hz)}$',fontsize=20)  
plt.ylabel('$\mathit{Magnification \ Factor \ (Ds)}$', fontsize=20)
```

```python
r = np.linspace(0,2, 1000)
plt.figure(figsize=(13.5,9)) 
plt.grid(color='r', linestyle='-', linewidth=0.2)
naturalFrequency = 2


def plotPhaseFunction(x):
    phaseAngle = np.arctan2((2*r*x),(1-np.square(r)))
    plt.plot(r, phaseAngle, label="$\mathit{Resistor \ Value \ }$"+str(x))

    
for x in {0,0.1,0.15,0.3,0.2,0.25,0.5,0.6,0.9,1}:
    plotPhaseFunction(x)
    plt.legend(loc="best")

plt.title('$\mathrm{Frequency \ Ratio \ Against \ Phase \ Angle}$',loc="center", fontsize=30)   
plt.xlabel('$\mathit{Frequency \ Ratio \ (Hz)}$',fontsize=20)  
plt.ylabel('$\mathit{Phase \ Angle (\degree)}$', fontsize=20)
```
