### Exeriment 7 Plotting

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(13.5,9)) 
Freq500, Freq500_Uncertainty, Voltage500, Voltage500_Uncertainty = np.loadtxt("E7 Data Vc - 500Ohm.csv", unpack=True, delimiter=",")
Freq1500, Freq1500_Uncertainty, Voltage1500, Voltage1500_Uncertainty = np.loadtxt("E7 Data Vc - 1500Ohm.csv", unpack=True, delimiter=",")


plt.errorbar(Freq500, Voltage500,xerr=Freq500_Uncertainty, yerr=Voltage500_Uncertainty,fmt='go', label="$\mathit{Resistor \ Value \ = \ 500 \Omega }$", linewidth=1.25)
plt.errorbar(Freq1500, Voltage1500,xerr=Freq1500_Uncertainty, yerr=Voltage1500_Uncertainty,fmt='ro', label="$\mathit{Resistor \ Value \ = \ 1500 \Omega }$", linewidth=1.25)


plt.legend(loc="best")
plt.grid(color='r', linestyle='-', linewidth=0.2)

xmajor_ticks = np.arange(0, 100000, 10000)                                               
ymajor_ticks = np.arange(0, 2.5, 0.5)  

plt.xticks(xmajor_ticks)                                           
plt.yticks(ymajor_ticks)   

plt.xlim([-1000,100000])


plt.title('$\mathrm{Peak \ to \ peak \ Voltage \ (Vc) \ Against \ Input \ Frequency}$',loc="center", fontsize=20)   #Add centred title as Roman font in Math Text 
plt.xlabel('$\mathit{Frequency \ (Hz)}$',fontsize=15)   #Add labels as Math Text using $ 
plt.ylabel('$\mathit{Peak \ to \ Peak \ Voltage \ over \ Capacitor \ (Vc)}$', fontsize=15)

plt.savefig("Voltage Response against Input Frequency.png", dpi=300)
```
